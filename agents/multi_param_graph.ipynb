# multinode graph apps


```python
from typing import TypedDict, Dict
from langgraph.graph import StateGraph
import math
from dataclasses import dataclass, field

# use data class for defaults
@dataclass
class AgentState:
    values : list[int] = field(default_factory=list)
    results : int = 0
    operation : str = "+"
```

```python
def process_vals(state: AgentState)-> AgentState:
    """ your result is calculated here"""
    if state.operation== "+" :
        state.results = sum(state.values)
    elif state.operation == "*":
        x =1
        state.results = math.prod(state.values)

    return state
```

```python
graph = StateGraph(AgentState)

graph.add_node("proccess_vals",process_vals)

graph.set_entry_point("proccess_vals")
graph.set_finish_point("proccess_vals")

app = graph.compile()
```

```python
from IPython.display import Image,display 
display(Image(app.get_graph().draw_mermaid_png()))
```
